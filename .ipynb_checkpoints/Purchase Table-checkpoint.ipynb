```python
# Setup
import pandas as pd

# Data file path
steam_200k = "Resources/steam-200k.csv"

# Read the steam_200k csv data and assign column names
steam_200k = pd.read_csv(steam_200k)
steam_200k.columns = ["user_id", "game_title", "play_purchase", "play_time", "0"]

# Drop rows containing 'purchase' since these don't have play time data
steam_200k = steam_200k[~steam_200k['play_purchase'].isin(['purchase'])]

# Display a preview of the data
steam_200k.head()
```

```python
# Read the titles csv data and merge with steam_200k csv
titles_df = pd.read_csv("Resources/titles.csv")
merged_df = pd.merge(titles_df, steam_200k, on = "game_title")

# Rename id column
merged_df.rename(columns = {'id':'game_id'}, inplace = True)

# Drop na values
merged_df.dropna()

# Display a preview of the data
merged_df.head()
```

```python
# Drop unneeded columns
df = merged_df.drop(['play_purchase', '0'], axis=1)

# Display a preview of the data
df.head()
```

```python
# Save the dataframe as a CSV file
df.to_csv('Purchase_Table.csv')
```
