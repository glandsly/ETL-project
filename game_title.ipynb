```python
import pandas as pd
```

```python
game_titles_df = pd.read_csv("resources/steam_games.csv")
```

```python
game_titles_df.head()
```

```python
trimmed_game_titles_df = game_titles_df.drop(columns=['url', 'types', 'recent_reviews', 'publisher', 'popular_tags', 'game_details', 'languages', 'achievements', 'desc_snippet', 'mature_content', 'minimum_requirements', 'recommended_requirements', 'discount_price'])
```

```python
trimmed_game_titles_df.head()
```

```python
top_games_df = pd.read_csv("resources/titles.csv")
```

```python
top_games_list = top_games_df['game_title'].to_list()
```

```python
top_games_list
```

```python
trimmed_game_titles_df.replace("ACE COMBATâ„¢ 7: SKIES UNKNOWN","ACE COMBAT 7: SKIES UNKNOWN")
trimmed_game_titles_df.replace("Rocket LeagueÂ®", "Rocket League")
trimmed_game_titles_df.replace("Sid Meierâ€™s Civilization VI", "Sid Meier's Civilization VI")
trimmed_game_titles_df.replace("Tom Clancy's Rainbow Six Siege", "Tom Clancy's Rainbow Six Siege")
dictionary = {"®":'', "™":''}
trimmed_game_titles_df.replace(dictionary, regex=True, inplace=True)
trimmed_game_titles_df['original_price'] = trimmed_game_titles_df['original_price'].str.replace('$', '')
trimmed_game_titles_df['all_reviews'].replace(',', '', regex=True, inplace=True)
```

```python
trimmed_game_titles_df = trimmed_game_titles_df[trimmed_game_titles_df.name.isin(top_games_list)]
```

```python
trimmed_game_titles_df = trimmed_game_titles_df.rename(columns={"name": "game_title"})
```

```python
trimmed_game_titles_df = trimmed_game_titles_df.sort_values('game_title')
```

```python
trimmed_game_titles_df = trimmed_game_titles_df.dropna()
```

```python
trimmed_game_titles_df.reset_index(drop=True, inplace=True)
```

```python
trimmed_game_titles_df
```

```python
trimmed_game_titles_df['all_reviews'] = trimmed_game_titles_df['all_reviews'].str.extract(pat = '\(([^)]+)')
```

```python
trimmed_game_titles_df['all_reviews'] = trimmed_game_titles_df['all_reviews'].astype(int)
```

```python
trimmed_game_titles_df = trimmed_game_titles_df.rename(columns={"all_reviews": "review_count"})
```

```python
trimmed_game_titles_df
```

```python
trimmed_game_titles_df.to_csv("resources/titles_trimmed.csv")
```
